```python
# Numerical features
num_features = ["age", "absences", "G1", "G2", "G3"]

# Plot for Math
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_features, 1):
    if col in math.columns:
        plt.subplot(2, 3, i)
        sns.boxplot(y=math[col])
        plt.title(f"Math - {col}")
plt.tight_layout()
plt.show()

# Plot for Portuguese
plt.figure(figsize=(15, 8))
for i, col in enumerate(num_features, 1):
    if col in portuguese.columns:
        plt.subplot(2, 3, i)
        sns.boxplot(y=portuguese[col])
        plt.title(f"Portuguese - {col}")
plt.tight_layout()
plt.show()
```

Boxplot Analysis (Outliers)

Math Dataset

Age: Concentrated between 15–19 years; few outliers at 21–22.

Absences: Most students have <10, but strong outliers exist with 70+ absences.

G1, G2, G3 (Grades): Mostly between 8–15. Some low outliers (0–3), no extreme high outliers.

Portuguese Dataset

Age: Similar to Math, mostly 15–19 years; rare outliers at 21–22.

Absences: Most <10, but outliers up to 30+ absences.

G1, G2, G3 (Grades): Higher on average than Math. Some low-grade outliers (0–5). Final grade (G3) is more compact and less variable than in Math.


Classification Task – Alcohol Consumption

```python
for df in [math, portuguese]:
    df['alc_level'] = ((df['Dalc'] + df['Walc']) / 2).apply(lambda x: 'High' if x >= 3 else 'Low')
```

```python
print(math['alc_level'].value_counts())
print(portuguese['alc_level'].value_counts())
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.countplot(data=math, x='alc_level', ax=axes[0], palette="Set2")
axes[0].set_title("Math Dataset - Alcohol Consumption")
axes[0].set_ylabel("Count")

sns.countplot(data=portuguese, x='alc_level', ax=axes[1], palette="Set2")
axes[1].set_title("Portuguese Dataset - Alcohol Consumption")
axes[1].set_ylabel("Count")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.countplot(data=math, x='Dalc', hue='alc_level', ax=axes[0], palette="coolwarm")
axes[0].set_title("Math - Weekday Alcohol Consumption (Dalc)")

sns.countplot(data=math, x='Walc', hue='alc_level', ax=axes[1], palette="coolwarm")
axes[1].set_title("Math - Weekend Alcohol Consumption (Walc)")

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14,5))

sns.countplot(data=portuguese, x='Dalc', hue='alc_level', ax=axes[0], palette="coolwarm")
axes[0].set_title("Math - Weekday Alcohol Consumption (Dalc)")

sns.countplot(data=portuguese, x='Walc', hue='alc_level', ax=axes[1], palette="coolwarm")
axes[1].set_title("Math - Weekend Alcohol Consumption (Walc)")

plt.show()
```

```python
features = ["absences", "activities", "romantic", "famsup"]

for feat in features:
    plt.figure(figsize=(6,4))
    sns.countplot(data=math, x=feat, hue="alc_level", palette="Set1")
    plt.title(f"Math - Alcohol vs {feat}")
    plt.show()
```

```python
grade_features = ["G1", "G2", "G3"]

for grade in grade_features:
    plt.figure(figsize=(6,4))
    sns.boxplot(data=math, x="alc_level", y=grade, palette="coolwarm")
    plt.title(f"Math - {grade} vs Alcohol Consumption")
    plt.show()
```
